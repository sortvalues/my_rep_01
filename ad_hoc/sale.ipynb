```python
import pandas as pd
```

```python
url='https://docs.google.com/spreadsheets/d/1jFCGxNP5B2_Gcg_VbZO-LZIsjZrAA9IF/edit?usp=sharing&ouid=116045163522511725897&rtpof=true&sd=true'
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_excel(url)
```

```python
df = df.drop(['Unnamed: 1','Unnamed: 2', 'Unnamed: 3', 'Unnamed: 5','Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 15'], axis = 1)
df = df.drop(index = [0,1,2,3,4,5])
df = df.rename(columns = {'Прайс-лист на 14 декабря 2024 г.' : 'id',
                     'Unnamed: 4' : 'product',
                     'Unnamed: 13' : 'price',
                     'Unnamed: 14' : 'stock'})
```

```python
df = df[df['stock'].isnull() != True]
df.price = round(df.price.astype(float) / 2)
df
```

```python
df.to_excel('price.xlsx', index = False)
```
